```python
import pandas as pd
import names
```

```python
file_to_load1= '../Resource/dataset/csv/participant-overview.xlsx'
overview=pd.read_excel(file_to_load1, header= 1)
overview.head()
```

```python
#Dropping unnecessary columns
overview_1= overview.drop(columns={
    'Height', 'A or B person', 'Max heart rate', 'Date', 'Minutes', 'Seconds', 'Stride walk', 'Stride run'
})
overview_1
```

```python
# Rename column
overview_2= overview_1.rename(columns={'Participant ID' : 'Participant_ID'})
overview_2
```

```python
# Generating random names by python 
last_name=[]
for i in range(16):
    addname = names.get_last_name()
    last_name.append(addname)
last_name
```
