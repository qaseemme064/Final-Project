```python
import pandas as pd
```

```python
file_to_load1= '../Resource/dataset/csv/not_cleaned_csv/p16/exercise.csv'
df1=pd.read_csv(file_to_load1)
df1.head()
```

```python
# Drop unnecessary columns
#df1= df1.drop(['logId','activityName', 'activityTypeId', 'activityLevel', 'logType', 'manualValuesSpecified', 
               #'heartRateZones', 'lastModified', 'startTime', 'originalDuration', 'elevationGain',
              #'hasGps', 'shouldFetchDetails', 'distanceUnit', 'source', 'tcxLink',
              #'speed', 'pace', 'vo2Max', 'duration'], axis = 1)
#df1.head()
df1= df1.loc[:, ['averageHeartRate','calories', 'activeDuration', 'steps', 'originalStartTime']]
df1.head()
```

```python
# Converting time in mins
df1['Active_Duration_in_mins'] = df1['activeDuration']/60000

# Seperating Date and Time
df1['Dates'] = pd.to_datetime(df1['originalStartTime']).dt.date
df1['Time'] = pd.to_datetime(df1['originalStartTime']).dt.time
df1.head()
```

```python
# Dropping unnecessary Columns
df1= df1.drop(['activeDuration', 'originalStartTime'], axis=1)
df1.head()
```

```python
# Renaming columns
rename_df = df1.rename(columns= {
    'averageHeartRate' : 'Avg_ActiveHeartRate',
    'calories' : 'Calories',
    'steps' : 'Steps',    
})
print(rename_df.shape)
rename_df.head()
```

```python
# Dropping Duplicates
rename_df=rename_df.drop_duplicates()
print(rename_df.shape)
```

```python
# Check for null values
rename_df.isnull().sum()
```

```python
# Check for notnull values
rename_df.notnull().sum()
```

```python
# Rearrange DataFrame
exercise_df = rename_df[['Dates', 'Time', 'Calories', 'Steps', 'Active_Duration_in_mins',
                         'Avg_ActiveHeartRate']]
print(exercise_df.shape)
exercise_df.head()
```

# For Calories Table

```python
# Takingout calories from dataFrame
calories  = exercise_df.loc[:, ['Dates', 'Calories']]
print(calories.shape)
calories.head()
```

```python
# adding same date 
calories_added = calories.groupby('Dates' , as_index=False).sum()
print(calories_added.shape)
calories_added.head(10)
```

```python
# Adding participant ID
participant_id = calories_added.insert(0, 'Participant_ID', 'p16')
```

```python
# Calories DataFrame
calories_df=pd.DataFrame(calories_added)
print(calories_df.shape)
calories_df.head(10)
```

```python
# saving calories file to csv
file_to_save = calories_df.to_csv(r'../Resource/dataset/csv/cleaned_csv/p16/calories_p16.csv', index = False)
```

# For Distance Table

```python
file_to_load1= '../Resource/dataset/csv/not_cleaned_csv/p16/distance.csv'
distance=pd.read_csv(file_to_load1)
distance.head()
```

```python
# checking for null and not null values
distance.isnull().sum()
distance.notnull().sum()
```

```python
# Seperating Date and Time and dropping dateTime Column
distance['Dates'] = pd.to_datetime(distance['dateTime']).dt.date
distance = distance.drop(columns={'dateTime'}, axis=1)
distance.head()
```

```python
distance = distance.rename(columns={
    'value' : 'Distance' 
})
distance = distance[['Dates', 'Distance']]
distance.head()
```

```python
# Dropping Dates that have NaN values
dates_drop_NaN=distance.dropna(subset=['Distance'])
print(dates_drop_NaN.shape)
dates_drop_NaN
```

```python
# Adding same date values if there is any
distance_added = dates_drop_NaN.groupby('Dates' , as_index=False).sum()
print(distance_added.shape)
distance_added
```

```python
# Getting only date columns and dropping duplicate dates
distance_drop = distance.drop('Distance', axis = 1)
distancevalue = distance_drop.drop_duplicates(subset=["Dates"])
print(distancevalue.shape)
distancevalue.head()
```

```python
# Merging dates and distance and sorted date wise
distance_merged=pd.merge(distance_added,distancevalue, how='outer' , on='Dates')
distance_sorted= distance_merged.sort_values('Dates')
print(distance_sorted.shape)
distance_sorted.head()
```

```python
# Adding participant ID
participant_id = distance_sorted.insert(0, 'Participant_ID', 'p16')
```

```python
# Distance DataFrame
distance_df=pd.DataFrame(distance_sorted)
print(distance_df.shape)
distance_df.head(10)
```

```python
distance_df['Distance']= distance_df['Distance'].fillna(0)
distance_df['Distance']= distance_df['Distance']/100000
print(distance_df.shape)
distance_df.head()
```

```python
# saving calories file to csv
file_to_save3 = distance_df.to_csv(r'../Resource/dataset/csv/cleaned_csv/p16/distance_p16.csv', index = False)
```

# Resting Heart Rate

```python
file_to_load1= '../Resource/dataset/csv/not_cleaned_csv/p16/resting_heart_rate.csv'
resting_heart_rate=pd.read_csv(file_to_load1)
print(resting_heart_rate.shape)
resting_heart_rate.head()
```

```python
# Splitting value column
resting_heart_rate[['date','value', 'error']] = resting_heart_rate['value'].str.split(',',expand=True)
resting_heart_rate
```

```python
# Further splitting value column
resting_heart_rate[['rem','bpm']] = resting_heart_rate['value'].str.split(':',expand=True)
resting_heart_rate
```

```python
# Dropping unnecessary column
cleaned_df = resting_heart_rate.drop(columns={
    'value', 'date', 'error', 'rem'
})
cleaned_df
```

```python
# Renaming Column
cleaned_rename = cleaned_df.rename(columns={
    'dateTime': 'Dates',
    'bpm': 'Resting_beats_per_min'
})
cleaned_rename
```

```python
# Adding participant ID
participant_id = cleaned_rename.insert(0, 'Participant_ID', 'p16')
```

```python
# Resting Heart Rate DataFrame
resting_heartrate_df=pd.DataFrame(cleaned_rename)
print(cleaned_rename.shape)
resting_heartrate_df.head(10)
```

```python
# saving resting heart rate file to csv
file_to_save7 = resting_heartrate_df.to_csv(r'../Resource/dataset/csv/cleaned_csv/p16/resting_heartrate_p16.csv', index = False)
```
