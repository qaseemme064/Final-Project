```python
import numpy as np
import pandas as pd
```

```python
file_to_load1= '../Resource/dataset/csv/for_machine_learning/not_cleaned_csv/p15/exercise.csv'
df1=pd.read_csv(file_to_load1)
df1.head()
```

```python
# Drop unnecessary columns
#df1= df1.drop(['logId','activityName', 'activityTypeId', 'activityLevel', 'logType', 'manualValuesSpecified', 
               #'heartRateZones', 'lastModified', 'startTime', 'originalDuration', 'elevationGain',
              #'hasGps', 'shouldFetchDetails', 'distanceUnit', 'source', 'tcxLink',
              #'speed', 'pace', 'vo2Max', 'duration'], axis = 1)
#df1.head()
df1= df1.loc[:, ['averageHeartRate','activeDuration', 'originalStartTime']]
df1.head()
```

```python
# Converting time in mins
df1['Active_Duration_in_mins'] = df1['activeDuration']/60000
df1= df1.drop(['activeDuration'], axis=1)
df1.head()
```

```python
# Renaming columns
rename_df = df1.rename(columns= {
    'averageHeartRate' : 'Avg_ActiveHeartRate',
    'originalStartTime' : 'datetime',
})
print(rename_df.shape)
rename_df.head()
```

```python
rename_df.isnull().sum()
```

```python
rename_df.notnull().sum()
```

# Avg Active Heart rate

```python
avg_active_rate  = rename_df.loc[:, ['datetime', 'Avg_ActiveHeartRate']]
print(avg_active_rate.shape)
avg_active_rate.head()
```

```python
avg_active_rate=avg_active_rate.dropna()
print(avg_active_rate.shape)
avg_active_rate.head()
```

```python
# saving sleep file to csv
file_to_save1 = avg_active_rate.to_csv(r'../Resource/dataset/csv/for_machine_learning/cleaned_csv/p15/avg_activeHeartRAte_p15.csv', index = False)
```

```python
# Changing datetime to datetime index
avg_active_rate['datetime'] = pd.to_datetime(avg_active_rate['datetime'])

#Converting minutes to hours by using groupby
avg_active_rate_p15 = avg_active_rate.groupby([pd.Grouper(freq='H', key='datetime')]).sum().reset_index()
avg_active_rate_p15
```

```python
#set datetime as index
avg_active_rate_p15 = avg_active_rate_p15.set_index(avg_active_rate_p15.datetime)

#drop datetime column
avg_active_rate_p15.drop('datetime', axis=1, inplace=True)

#create hour, day and month variables from datetime index

avg_active_rate_p15['hour'] = avg_active_rate_p15.index.hour
avg_active_rate_p15['day'] = avg_active_rate_p15.index.day
avg_active_rate_p15['month'] = avg_active_rate_p15.index.month
avg_active_rate_p15
```

```python
#Saving file for machine learning
file_to_save_avg_active_rate = avg_active_rate_p15.to_csv(r'../Resource/dataset/csv/for_machine_learning/for_predictions_csv/p15/avg_active_heart_rate_p15.csv')
```

# Resting Heart Rate

```python
file_to_load1= '../Resource/dataset/csv/for_machine_learning/not_cleaned_csv/p15/resting_heart_rate.csv'
resting_heart_rate=pd.read_csv(file_to_load1)
print(resting_heart_rate.shape)
resting_heart_rate.head()
```

```python
# Splitting value column
resting_heart_rate[['date','value', 'error']] = resting_heart_rate['value'].str.split(',',expand=True)
resting_heart_rate

# Further splitting value column
resting_heart_rate[['rem','bpm']] = resting_heart_rate['value'].str.split(':',expand=True)

cleaned_df = resting_heart_rate.drop(columns={
    'value', 'date', 'error', 'rem'
})
```

```python
cleaned_rename = cleaned_df.rename(columns={
    'dateTime': 'datetime',
    'bpm': 'Resting_beats_per_min'
})
cleaned_rename
```

```python
file_to_save7 = cleaned_rename.to_csv(r'../Resource/dataset/csv/for_machine_learning/cleaned_csv/p15/resting_HeartRate_p15.csv', index = False)
```

```python
cleaned_rename['Resting_beats_per_min'] = cleaned_rename['Resting_beats_per_min'].astype(float, errors = 'raise')

# Changing datetime to datetime index
cleaned_rename['datetime'] = pd.to_datetime(cleaned_rename['datetime'])

#Converting minutes to hours by using groupby
resting_heart_rate_p15 = cleaned_rename.groupby([pd.Grouper(freq='D', key='datetime')]).sum().reset_index()
resting_heart_rate_p15
```

```python
#set datetime as index
resting_heart_rate_p15 = resting_heart_rate_p15.set_index(resting_heart_rate_p15.datetime)

#drop datetime column
resting_heart_rate_p15.drop('datetime', axis=1, inplace=True)

#create hour, day and month variables from datetime index
#resting_heart_rate_p15['hour'] = resting_heart_rate_p15.index.hour
resting_heart_rate_p15['day'] = resting_heart_rate_p15.index.day
resting_heart_rate_p15['month'] = resting_heart_rate_p15.index.month
resting_heart_rate_p15
```

```python
#file for machine learning
file_to_save_resting_heart_rate = resting_heart_rate_p15.to_csv(r'../Resource/dataset/csv/for_machine_learning/for_predictions_csv/p15/resting_heart_rate_p15.csv')
```
